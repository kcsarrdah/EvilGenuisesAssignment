```python
import matplotlib.pyplot as plt
import pyarrow.parquet as pq
import numpy as np
import pandas as pd
```

```python
#defining the boundry to check 

# Z-axis boundaries
z_min = 285
z_max = 421

# Coordinates for each boundary
boundary_13 = [[-1735, 250], [-2024, 398], [z_min, z_max]]
boundary_14 = [[-2024, 398], [-2806, 742], [z_min, z_max]]
boundary_15 = [[-2806, 742], [-2472, 1233], [z_min, z_max]]
boundary_16 = [[-2472, 1233], [-1565, 580], [z_min, z_max]]

# Generate all the boundary points
boundary_points = []
for x in [boundary_13[0], boundary_14[0], boundary_15[0], boundary_16[0]]:
    for y in [boundary_13[1], boundary_14[1], boundary_15[1], boundary_16[1]]:
        for z in [boundary_13[2]]:
            boundary_points.append((x[0], y[0], z[0]))  # Lower-left vertex
            boundary_points.append((x[0], y[0], z[1]))  # Upper-left vertex
            boundary_points.append((x[0], y[1], z[0]))  # Lower-right vertex
            boundary_points.append((x[0], y[1], z[1]))  # Upper-right vertex
            boundary_points.append((x[1], y[0], z[0]))  # Lower-front vertex
            boundary_points.append((x[1], y[0], z[1]))  # Upper-front vertex
            boundary_points.append((x[1], y[1], z[0]))  # Lower-back vertex
            boundary_points.append((x[1], y[1], z[1]))  # Upper-back vertex

# Extract x, y, z coordinates from boundary points
x_coords = [point[0] for point in boundary_points]
y_coords = [point[1] for point in boundary_points]
z_coords = [point[2] for point in boundary_points]

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the boundary points
ax.scatter(x_coords, y_coords, z_coords)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Boundary Points')

# Show the plot
plt.show()
```

```python
#parsing the parquet file and storing all the data in a list

# Provide the file path of the Parquet file
parquet_file_path = 'game_state_frame_data.parquet'

# Open the Parquet file
parquet_file = pq.ParquetFile(parquet_file_path)

# Get the number of row groups in the Parquet file
num_row_groups = parquet_file.num_row_groups

# Initialize the list to store the data
data = []

# Iterate over each row group
for row_group in range(num_row_groups):
    # Read the row group as a pandas DataFrame
    df = parquet_file.read_row_group(row_group).to_pandas()

    # Iterate over each row and append it to the data list
    for _, row in df.iterrows():
        data.append(row)

# Display the first 10 values in the data list
for i in range(3):
    print(data[i], end='\n')
    
```

```python

def check_boundary(file_path, vertices, z_min, z_max):
    # Read the Parquet file into a pandas DataFrame
    df = pd.read_parquet(file_path)

    # Initialize the count variables and the list for matching entries
    total_entries = 0
    matching_entries = []
    matching_entries_data = []

    # Iterate over the rows in the DataFrame
    for i, row in df.iterrows():
        total_entries += 1

        # Check the boundary conditions
        within_z_boundary = (row['z'] >= z_min) and (row['z'] <= z_max)
        within_volume = []

        for vertices_x, vertices_y in vertices.values():
            within_volume.append((row['x'] >= vertices_x[0]) and (row['x'] <= vertices_x[1]) and
                                 (row['y'] >= vertices_y[0]) and (row['y'] <= vertices_y[1]))

        within_volume = any(within_volume)

        # Check if the row falls within the specified volume
        if within_z_boundary and within_volume:
            matching_entries.append(total_entries)
            matching_entries_data.append(row)

    # Calculate the percentage of matching entries
    percentage = (len(matching_entries) / total_entries) * 100

    return matching_entries, matching_entries_data, percentage

# Provide the file path of the Parquet file
parquet_file_path = 'game_state_frame_data.parquet'

# Set the vertices and z-axis boundary values
vertices = {
    13: ([-1735, 250], [-2024, 398]),
    14: ([-2024, 398], [-2806, 742]),
    15: ([-2806, 742], [-2472, 1233]),
    16: ([-2472, 1233], [-1565, 580])
}
z_min = 285
z_max = 421

# Call the function to check the boundaries and store the matching entries
matching_entries, matching_entries_data, percentage = check_boundary(parquet_file_path, vertices, z_min, z_max)

# Print the matching entries
print("Matching Entries:", len(matching_entries))

# Print the matching entries data
for i in range(min(4, len(matching_entries_data))):
    print(matching_entries_data[i])

# Print the percentage of data falling within the specified category
print("Percentage:", percentage)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Initialize a nested dictionary to store the player entries for each round
round_player_entries = {}

# Iterate over the data array
for row in data:
    round_num = row['round_num']
    if 16 <= round_num <= 30 and row['area_name'] == 'BombsiteB' and row['team'] == 'Team2' and row['side'] == 'T':
        player = row['player']
        if round_num not in round_player_entries:
            round_player_entries[round_num] = {}
        
        if player not in round_player_entries[round_num]:
            round_player_entries[round_num][player] = []
        
        round_player_entries[round_num][player].append(row)

# Sort the round numbers in ascending order
sorted_rounds = sorted(round_player_entries.keys())

# Prepare data for the chart
valid_rounds = []
unique_players_counts = []
round_seconds = []  # List to store the seconds values for each entry in each valid round

for round_num in sorted_rounds:
    player_entries = round_player_entries[round_num]
    
    # Check if the round has a total of two SMGs, two rifles, or one SMG and one rifle
    num_smgs = 0
    num_rifles = 0
    round_entry_seconds = []  # List to store the seconds values for each entry in the round
    
    for player, entries in player_entries.items():
        for entry in entries:
            seconds = entry['seconds']
            round_entry_seconds.append(seconds)
            
            inventory = entry['inventory']
            if inventory is not None:
                for item in inventory:
                    if item['weapon_class'] == 'SMG':
                        num_smgs += 1
                    elif item['weapon_class'] == 'Rifle':
                        num_rifles += 1
    
    if num_smgs >= 2 or num_rifles >= 2 or (num_smgs + num_rifles >= 2):
        valid_rounds.append(round_num)
        unique_players_counts.append(len(player_entries))
        round_seconds.append(round_entry_seconds)

# Create a bar chart
plt.bar(valid_rounds, unique_players_counts)
plt.xlabel('Round')
plt.ylabel('Number of Unique Players')
plt.title('Player Entries in Bombsite B for Each Round')
plt.xticks(valid_rounds)
plt.show()

# Print the seconds values for each entry in each valid round
for i, round_num in enumerate(valid_rounds):
    print(f"Round {round_num}: Seconds Values = {round_seconds[i]}")
    
# Flatten the list of seconds values
flattened_seconds = [seconds for round_seconds_list in round_seconds for seconds in round_seconds_list]

# Calculate the average value of the seconds values
average_seconds = np.mean(flattened_seconds)
print(f"Average Seconds: {average_seconds}")
```

```python
# Extract the X and Y coordinates from the DataFrame
x_coords = df_filtered['x']
y_coords = df_filtered['y']

# Create a 2D histogram of the coordinates
heatmap, xedges, yedges = np.histogram2d(x_coords, y_coords, bins=25)

# Plot the heatmap
plt.imshow(heatmap.T, origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], cmap='hot')
plt.colorbar(label='Frequency')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Heatmap of X-Y Coordinates')

# Find the coordinates with frequency above 250
above_threshold_coords = []
for i, x in enumerate(xedges[:-1]):
    for j, y in enumerate(yedges[:-1]):
        if heatmap[i, j] > 250:
            above_threshold_coords.append((x, y))

# Print the coordinates with frequency above 250
print("Coordinates with frequency above 250:")
for coord in above_threshold_coords:
    print(coord)

plt.show()
```
